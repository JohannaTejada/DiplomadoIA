**Problema: Evaluación de la tasa de deterioro cognitivo**

**Contexto**

Un grupo de investigadores quiere estudiar la progresión del deterioro cognitivo en pacientes con Alzheimer. Para esto, se usa el puntaje MMSE (Mini-Mental State Examination), una prueba que evalúa el estado cognitivo general del paciente. Se plantea relacionar este puntaje con el tiempo transcurrido desde el diagnóstico de la enfermedad.

Columnas del archivo .csv

Variable	Tipo de dato	Descripción
ID_Paciente	string	Identificador único
Edad	entero	Edad del paciente
MMSE	entero	Puntuación del Mini-Mental State Examination (0–30)
MoCA	entero	Puntuación del Montreal Cognitive Assessment (0–30)
Volumen_Hipocampo	flotante	Medido por neuroimagen (cm³ o mm³)
Años_Diagnóstico	entero	Tiempo desde el diagnóstico
Nivel_Actividad_Física	flotante	Horas semanales de actividad física
Índice_Masa_Corporal	flotante	BMI del paciente
Porcentaje_Pérdida_Memoria	flotante	Estimación autoinformada o médica del deterioro (0–100%)

**Problema**

Cada paciente tiene una puntuación MMSE y un valor que indica cuántos años han pasado desde su diagnóstico. Se desea calcular una nueva medida llamada Tasa de Deterioro Cognitivo, que se define como el cociente entre el puntaje MMSE y los años desde el diagnóstico.

Sin embargo, hay pacientes que apenas han sido diagnosticados (0 años), por lo que en estos casos la tasa se definirá como igual al valor MMSE (para evitar divisiones por cero).

Solución esperada

El estudiante debe imprimir esta 'Tasa de Deterioro Cognitivo'. Para cada paciente, esta tasa debe calcularse dividiendo su puntaje MMSE entre el número de años desde el diagnóstico.

En los casos donde el número de años sea igual a cero, se debe asignar directamente el valor de MMSE como tasa.

El resultado debe ser mostrado para cada paciente, indicando su ID, puntaje MMSE, años desde el diagnóstico y la tasa calculada, con un formato claro y legible.

**Retos**

Contar cuántos pacientes tienen una tasa de deterioro menor a 2.0.
Calcular el promedio general de la tasa de deterioro en toda la muestra.
Crear un nuevo diccionario con los pacientes que tienen tasas bajas y podrían considerarse en riesgo alto de deterioro.


````python
import csv
import urllib.request

# URL del dataset
url = 'https://raw.githubusercontent.com/juliandariogiraldoocampo/ia_taltech/refs/heads/main/pacientes_alzheimer_datos_sinteticos.csv'

# Descargar y leer el archivo CSV como lista de diccionarios
response = urllib.request.urlopen(url)

# Leer las lineas del archivo y decodificar los bytes a texto (UTF-8)
lines = [l.decode('utf-8') for l in response.readlines()]

# Convertir el contenido de la lista (archivo csv) en una lista de diccionarios
data = list(csv.DictReader(lines))

pacientes = {}
for row in data:
    pacientes[row['ID_Paciente']] = row

# Calcular y mostrar la Tasa de Deterioro Cognitivo
for cada_paciente in pacientes:
    paciente = pacientes[cada_paciente]

    mmse = float(paciente['MMSE'])
    annos = int(paciente['Annos_Diagnostico'])

    if annos ==  0:
        tasa = round(mmse, 2)
    else:
        tasa = round(mmse / annos, 2)

    paciente['Tasa_Deterioro'] = tasa

    print(f'Paciente {registro} → MMSE: {mmse}, Años Dx: {annos}, Tasa: {tasa}')
````

**9. Etiquetado de riesgo**

Contexto:

Se necesita generar etiquetas de riesgo a partir de variables básicas.

Problema:

Pide al usuario edad y temperatura. Si edad > 65 y temperatura > 38.0, muestra "ALTO RIESGO", si no, muestra "RIESGO CONTROLADO".

Solución esperada:

Condicional anidado con operador and.

Reto:

````python
import csv
import urllib.request

cont = 0

# URL del dataset
url = 'https://raw.githubusercontent.com/JohannaTejada/DiplomadoIA/refs/heads/main/Temperatura.csv'

# Descargar y leer el archivo CSV como lista de diccionarios
response = urllib.request.urlopen(url)

# Leer las lineas del archivo y decodificar los bytes a texto (UTF-8)
lines = [l.decode('utf-8-sig') for l in response.readlines()]

# Convertir el contenido de la lista (archivo csv) en una lista de diccionarios
data = list(csv.DictReader(lines))

pacientes = {}

# Con esto creo un diccionario VACÍO
for row in data:

# Recorro la data fila(row) por fila(row)
    pacientes [row['ID']] = row

# al diccionario en la Posición(fila) le llevo la fila entera

print(' '*30,'INFORME')
print('#'*70)

# Calcular y mostrar la Etiqueta de RIESGO
for cada_paciente in pacientes:
    paciente = pacientes[cada_paciente]

    edad = int(paciente['Edad'])
    temperatura = float(paciente['Temperatura'])

    if edad > 65 and temperatura > 38:
        etiqueta = "ALTO RIESGO"
        cont = cont + 1
    else:
        etiqueta = "RIESGO CONTROLADO"

    paciente['Etiqueta'] = etiqueta

    print(f'Paciente {cada_paciente}, Edad {edad}, Temperatura {temperatura}°C →  {etiqueta}')
    print('-'*70)
print(' '*30,'PACIENTES CON ALTO RIESGO')
print(' '*20, f'La cantidad de pacientes con ALTO RIEGO SON {cont}')
````

````python
import numpy as np
import pandas as pd

# Configurar semilla para reproductibilidad
np.random.seed(42)

# Inicializar Variables
# Número de Observaciones/registros
n = 100
# Probabilidad de cada opción
prob =  [0.5, 0.5]
opciones = ["Masculino", "Femenino"]

# Generar datos sintéticos de edad
# Media, Desv.Estándar, Número de Observaciones
edad = np.random.normal(70, 8, n)

# Generar datos sintéticos de género con Probabilidad 50% cada uno
genero = np.random.choice(opciones, size=n, p=prob)

# Si el género afecta el puntaje de memoria
# Condición
# Si... Hombres con media 26
# Sino... Mujeres con media 24
memoria = np.where(genero == "Masculino",
                  np.random.normal(26, 5, n),
                  np.random.normal(24, 5, n))

# Crear el Dataframe con las listas
df_sintetico = pd.DataFrame({
    'Edad': edad,
    'Puntaje_memoria': memoria,
    'Género': genero
})

df_sintetico.head()
````

<font size=5>📢 Claves para usar np.random:

<font size=4>**Distribución simétrica:**
La mayoría de los datos se concentran cerca de la media, y los extremos son menos probables.

<font size=4>**Uso típico:**
Modelar características como alturas, puntajes de tests, errores de medición, etc.

<font size=4>**¿Por qué no usar `np.random.rand()`?**

<font size=4>`rand()` genera datos uniformes (todos los valores tienen igual probabilidad), mientras que `normal()` refleja fenómenos reales donde los valores centrales son más comunes.

## **<font size = 6>✅ Reto 3. Normalización de Datos**

**🔍 Introducción a la Normalización de Datos**

En el contexto del machine learning, la normalización es un paso crítico en el preprocesamiento de datos que consiste en ajustar las escalas de las variables numéricas para que todas compartan un rango o distribución similar.

**¿Por qué es importante?**

- **Evita sesgos en los modelos:** Algoritmos como K-Nearest Neighbors (KNN), Support Vector Machines (SVM) o Principal Component Analysis (PCA) son sensibles a las diferencias de magnitud entre variables. Por ejemplo, si una variable (como "ingresos anuales") está en miles y otra (como "edad") en decenas, el modelo podría sobrevalorar la primera simplemente por su escala numérica.

- **Mejora la convergencia:** En modelos basados en gradiente descendente (como redes neuronales), la normalización acelera el entrenamiento al equilibrar la velocidad de actualización de los pesos.

- **Facilita comparaciones**: Permite interpretar coeficientes o importancias de características de manera más justa.

**Casos de uso común**

- Variables como glucosa (mg/dL), presión arterial (mmHg) e índice de masa corporal (kg/m²) suelen estar en escalas muy distintas. Normalizarlas asegura que ninguna domine el modelo artificialmente.

> <font size=5>💡Para tener en cuenta:</font><br>
> *La normalización es una técnica fundamental para garantizar que los datos estén en condiciones óptimas antes de entrenar modelos, especialmente cuando las variables tienen unidades o rangos dispares.*<br>
<i>**Nota:** Las técnicas más usadas incluyen `StandardScaler` (media=0, desviación=1) y `MinMaxScaler` (escala [0, 1]). Elegir una u otra depende del algoritmo y la distribución de los datos.

````python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

########## CARGA DE DATOS
url = 'https://github.com/juliandariogiraldoocampo/ia_taltech/raw/refs/heads/main/Dataset%201%20_%20Pima%20Indians%20diabetes%20dataset%20(PIDD)_modificado.xlsx'

# Cargar la hoja sin asumir encabezados
df_raw = pd.read_excel(url, header=None) # En caso de no querer cargar los encabezados del archivo

# Fila 5 contiene los nombres de las columnas reales (índice 5)
columnas = df_raw.iloc[5, 1:].tolist()


########## LIMPIEZA
# Extraemos desde la fila 6 hasta la 774 (excluyendo las últimas 3 filas)
df = df_raw.iloc[6:774, 1:]  # <- NOTA: El índice 774 NO se incluye

# Asignamos nombres de columnas y reiniciamos índice
df.columns = columnas

df.reset_index(drop=True, inplace=True)

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

columnas_a_imputar = ['pregnancies',  'glucose',  'blood_pressure', 'skin_thickness', 'insulin', 'body_mass_index', 'diabetes_pedigree_function', 'age']


########## IMPUTACION
for col in columnas_a_imputar:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # asegurar tipo numérico

    df[col] = df[col].replace(0, np.nan)

# Crear el imputador
imputer = KNNImputer(n_neighbors=5)

# Aplicar imputación
df_imputado = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

df = df_imputado
````

## **<font size = 6>✅ Reto 4. Análisis Univariable**

El análisis univariable es el primer paso en la exploración de datos (EDA), donde se estudia una variable a la vez para entender su comportamiento, distribución y características clave. Este enfoque permite:

- Identificar patrones o tendencias en los datos.

- Detectar valores atípicos (outliers) o errores en la medición.

- Evaluar la dispersión y el centro de los datos (media, mediana, moda, rango, desviación estándar).

- Preparar los datos para etapas posteriores, como el modelado estadístico o de machine learning.

````python
# Estadísticas individuales más detalladas
from scipy.stats import skew, kurtosis

columna = 'glucose'
print(f"Asimetría (skew) de {columna}: ", skew(df[columna], nan_policy='omit'))
print(f"Curtosis de {columna}: ", kurtosis(df[columna], nan_policy='omit'))
````

**🔍 ¿Qué es la Curtosis?**

La curtosis (o kurtosis) es una medida estadística que cuantifica qué tan "pesadas" o ligeras son las colas de una distribución comparada con la distribución normal (gaussiana). Indica la presencia de valores extremos (outliers) y la concentración de datos alrededor de la media. Se interpreta así:

- **Curtosis = 0**: La distribución tiene colas similares a la normal (mesocúrtica).

- **Curtosis > 0**: Colas más pesadas y pico más agudo (leptocúrtica → más outliers).

- **Curtosis < 0**: Colas más ligeras y pico más aplanado (platicúrtica → menos outliers).

**🤔 ¿Para qué sirve esto?**
- Identificar distribuciones no normales: Ayuda a decidir si aplicar transformaciones (ej: logarítmica).

- Preparar datos para modelos: Algoritmos como SVM o KNN son sensibles a asimetría y outliers.

- Detectar anomalías: Curtosis alta sugiere presencia de valores extremos.

> <font size = 6>💡</font>*Transformamos datos no normales para que los modelos de machine learning funcionen mejor. Si los datos están muy cargados a un lado o tienen valores extremos, los algoritmos pueden cometer errores o volverse lentos. <br><br>
Usamos transformaciones sencillas (como el logaritmo o la raíz cuadrada) para **"suavizar" esos datos y que se parezcan más a una campana de Gauss**, lo que ayuda a los modelos a aprender patrones de manera más precisa.*

### <font size = 5> **Validación de normalidad**

#### 🔸 Test de Shapiro-Wilk:

Evalúa si una muestra proviene de una distribución normal.

<i><b>Uso ideal:</b> Para muestras pequeñas (típicamente < 5000 datos).</i>

````python
from scipy.stats import shapiro

stat, p = shapiro(df['glucose'])
print(f'Estadístico de Shapiro-Wilk: {stat:.4f}, valor p: {p:.4f}')

if p > 0.05:
    print('Distribución normal (no se rechaza H0)')
else:
    print('Distribución no normal (se rechaza H0)')
````

####🔸 Test de Pearson:

Combina asimetría (skew) y curtosis (kurtosis) para evaluar normalidad.

<i><b>Uso ideal:</b> Para muestras más grandes (> 20 datos, recomendable >50).</i>

````python
from scipy.stats import normaltest
stat, p = normaltest(df['glucose'])
print("p-value:", p)  # Si p < 0.05, rechaza normalidad
````

#### **Hipótesis nula (H0) vs. Hipótesis alternativa (H1)**

- **Hipótesis nula (H0):** Es la "posición por defecto" que asumimos inicialmente.

 Ejemplo: "Los niveles de glucosa siguen una distribución normal".

- **Hipótesis alternativa (H1):** Es lo que queremos probar (lo opuesto a H0).

 Ejemplo: "Los niveles de glucosa no siguen una distribución normal".
---
**Interpretación clave (independiente del test):**

1. **Si la evidencia (p-value) es débil (p > 0.05):**

 No podemos rechazar H0 (no hay pruebas suficientes para descartar la normalidad).

 No significa que H0 sea verdadera, solo que no hay datos contundentes en contra.

2. **Si la evidencia es fuerte (p ≤ 0.05):**

 Rechazamos H0 (hay suficiente apoyo para H1).

 Conclusión: Los datos se desvían significativamente de la normalidad.

**Analogía médica:**

- H0: "El paciente está sano".
- H1: "El paciente tiene una condición".
- p-value: Resultado de un examen diagnóstico.

 - p > 0.05: "No hay hallazgos concluyentes para diagnosticar la condición".

 - p ≤ 0.05: "Hallazgos sugieren la presencia de la condición".

> <font size = 4>🚨**Pilas con esto:** <br>
> - <font size = 4>H0/H1 son afirmaciones sobre la población, **no sobre la muestra**.
> - <font size = 4>Nunca se "prueba" H0, solo se evalúa si hay evidencia suficiente ara rechazarla.
> - <font size = 4>El umbral de 0.05 es convencional, no mágico (en medicina a veces e usa 0.01 para mayor rigor).
> - <font size = 4>¡Esto aplica a cualquier test de hipótesis, no solo a normalidad!

#### **Otro ejemplo de H0 y H1**

**Analogía del Detective (Juicio Penal)**
- **H0 (Hipótesis nula):** "El acusado es inocente" (como en "inocente hasta que se pruebe lo contrario").
- **H1 (Hipótesis alternativa):** "El acusado es culpable".
- **p-value:** La fuerza de la evidencia (huellas, testigos, ADN).
 - Si p > 0.05: Evidencia débil → No condenas (no rechazas H0: "inocente").
 - Si p ≤ 0.05: Evidencia sólida → Condenas (rechazas H0: "culpable").

**<font size = 4>🚨Ojo, esto es clave:**

- Nunca pruebas la inocencia (H0), solo buscas pruebas contra ella.
- Un veredicto de "no culpable" <font size = 5>≠</font> "inocente", solo que no hubo pruebas suficientes.

### 📊 **Tabla de decisión:**

Test de normalidad por tamaño de muestra

| Tamaño de la Muestra | Test Recomendado           | Razón                                                                 |
|-----------------------|----------------------------|-----------------------------------------------------------------------|
| **n < 50**           | `shapiro()` (Shapiro-Wilk) | Más sensible para detectar no-normalidad en muestras pequeñas.        |
| **50 ≤ n ≤ 5,000**   | `normaltest()` (D'Agostino-Pearson) | Combina *skew* y *kurtosis*; robusto para muestras medianas.          |
| **n > 5,000**        | Gráficos (QQ-plot, histograma) + `normaltest()` | Los tests pueden ser demasiado sensibles; prioriza visualización.     |

**Notas**:  
- Para **n < 20**, considera pruebas no paramétricas o transformaciones automáticas.  
- Si el p-value es **< 0.05**, rechaza normalidad (pero evalúa el impacto real en tu modelo).  

### 📊 **¿Qué es el p-value?**

El p-value (valor de p) es un número entre 0 y 1 que nos dice si los datos siguen una distribución normal (en forma de "campana de Gauss").

**Interpretación:**

- Si p-value > 0.05: Los datos probablemente son normales ("aceptamos" la normalidad).
- Si p-value ≤ 0.05: Los datos no son normales (hay evidencia para rechazar la normalidad).
---
🧪 Comparación con un Examen Médico
Imaginemos que estos tests son como pruebas diagnósticas:

**1. Shapiro-Wilk (para muestras pequeñas, <50 datos):**
- Similar a un test rápido de glucosa. Es preciso con pocas muestras, pero pierde sensibilidad si hay muchos datos.
- Ejemplo: Si p = 0.03 → "Hay evidencia de que los datos no son normales" (como un test positivo para una anomalía).

**2. D'Agostino-Pearson (para muestras más grandes):**
- Similar a un panel metabólico completo. Evalúa dos aspectos:
 - Asimetría (si la curva está desplazada a izquierda/derecha).
 - Curtosis (si las colas son más gruesas o planas que lo normal).
- Ejemplo: Si p = 0.20 → "No hay evidencia suficiente para decir que los datos no son normales" (como un resultado negativo en un examen).

📊 **¿Por qué importa en investigación clínica?**
Muchos análisis estadísticos (como el t-test o ANOVA) asumen normalidad. Si el p-value ≤ 0.05:
- Alternativas: Usar pruebas no paramétricas (Mann-Whitney, Kruskal-Wallis).
> Las pruebas no paramétricas son métodos estadísticos que no asumen una distribución normal en los datos y se usan **cuando los datos son ordinales** o **no cumplen los requisitos de las pruebas tradicionales**.
> Ejemplos comunes:
> - Prueba de Mann-Whitney (comparar 2 grupos)
> - Prueba de Kruskal-Wallis (comparar 3+ grupos)
> - Prueba de Wilcoxon (datos pareados)
> Ventaja clave: Funcionan bien con muestras pequeñas o datos atípicos.
- Transformar datos: Aplicar logaritmo para normalizar.

## **<font size = 5>✅ Reto 6. PCA - Análisis de Componentes Principales**

PCA es una técnica que simplifica datos complejos al combinar variables correlacionadas en nuevos ejes ("componentes principales"), ordenados por la varianza que explican. Dico de optra forma, combina variables correlacionadas (ej: glucosa/insulina) en componentes simplificados que capturan la esencia de los datos.

**Objetivos Clave**

- **Reducir dimensiones:**

 Transformar 10+ variables en solo 2-3 componentes sin perder información crítica.

- **Capturar varianza:**

 El primer componente (PC1) explica la mayor variabilidad; el segundo (PC2) la siguiente, y así sucesivamente.

- **Visualizar patrones:**

 Graficar datos en 2D/3D para identificar agrupaciones naturales (ej: pacientes sanos vs. diabéticos).

**¿Por qué usar PCA?**

- **Elimina redundancia:** Variables como glucosa e insulina (correlacionadas) se fusionan en un componente.

- **Optimiza modelos:** Menos dimensiones = menos sobreajuste + mayor velocidad.

- **Revela estructura oculta:** Agrupaciones no visibles en altas dimensiones.

 ><b><i>Ejemplo clínico:</b> <br>
 Si PC1 combina glu cosa, insulina y edad, podría interpretarse como un "factor de riesgo metabólico".</i>

 > ⚠️ Consideraciones
 > - Escala tus datos antes de aplicar PCA (usar StandardScaler).
 > - Interpretabilidad: Los componentes son combinaciones matemáticas, no variables médicas directas.

### 🔹 Paso 1: Estandarización de las variables

PCA requiere que las variables estén en la misma escala.

````python
from sklearn.preprocessing import StandardScaler

# Selección de variables numéricas
variables = ['glucose', 'blood_pressure', 'skin_thickness', 'insulin', 'body_mass_index', 'age']

# Estandarización
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[variables])
````

### 🔹 Paso 2: Aplicar PCA

````python
from sklearn.decomposition import PCA

# Inicializar PCA para dos componentes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Crear DataFrame con componentes
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
````

 Paso 3: Visualizar componentes principales

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df_pca, x='PC1', y='PC2')
plt.title('Visualización PCA - 2 Componentes Principales')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} varianza)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} varianza)')
plt.grid(True)
plt.show()
````

### 📊 Paso 4: Interpretación de varianza explicada

````python
# Porcentaje de varianza explicada por cada componente
print('Varianza explicada:', pca.explained_variance_ratio_)
````

❓❓❓<font size = 5>¿Cuántos componentes necesitas para explicar el 80% de la varianza?

````python
# PCA sin limitar componentes
pca_full = PCA()
pca_full.fit(X_scaled)

# Varianza explicada por cada componente
print(pca_full.explained_variance_ratio_)
````
