# <strong> Introducci√≥n a la IA y  Tipos de Aprendizaje </strong>

## <strong> üìúHistoria y Evoluci√≥n de la IA </strong>

La historia de la inteligencia artificial (IA) comenz√≥ en la antig√ºedad con mitos sobre seres artificiales inteligentes, pero su desarrollo moderno se remonta a mediados del siglo XX. En la d√©cada de 1940, la invenci√≥n de la computadora digital programable sent√≥ las bases para la IA. En 1950, Alan Turing public√≥ su art√≠culo "Computing Machinery and Intelligence", introduciendo el concepto del test de Turing, que eval√∫a la capacidad de una m√°quina para exhibir comportamiento inteligente indistinguible del de un ser humano.

En 1956, la conferencia en Dartmouth College marc√≥ el nacimiento formal de la IA como disciplina acad√©mica, liderada por John McCarthy, Marvin Minsky, Nathaniel Rochester y Claude Shannon. La IA ha experimentado avances significativos, incluyendo el desarrollo del aprendizaje autom√°tico y el aprendizaje profundo, as√≠ como per√≠odos de desilusi√≥n financiera conocidos como los "inviernos de la IA".

## <strong> ü´ß Diferencias entre IA, ML y Deep Learning </strong>

- **Inteligencia Artificial (IA):** Se refiere a la creaci√≥n de sistemas que pueden realizar tareas que normalmente requieren inteligencia humana. La IA abarca un amplio rango de t√©cnicas y disciplinas, desde el procesamiento del lenguaje natural hasta la visi√≥n artificial.

- **Machine Learning (ML):** Es un subconjunto de la IA que se enfoca en el desarrollo de algoritmos que permiten a las m√°quinas aprender de los datos sin ser programadas expl√≠citamente. El ML se utiliza para predecir resultados o clasificar datos.

- **Deep Learning (DL):** Es una rama del ML que utiliza redes neuronales artificiales para analizar datos complejos, como im√°genes y audio. El DL ha sido crucial en avances recientes en √°reas como la visi√≥n artificial y el procesamiento del lenguaje natural.

## <strong> üßëüèª‚Äçüè´Tipos de Aprendizaje </strong>

- **Aprendizaje Supervisado**: El aprendizaje supervisado es un paradigma donde el modelo es entrenado con datos etiquetados. Se utilizan entradas y sus correspondientes salidas deseadas para aprender una funci√≥n de mapeo.

    - Regresi√≥n Lineal para predecir valores cl√≠nicos en pacientes.

    - Clasificaci√≥n con Naive Bayes para detectar enfermedades en im√°genes m√©dicas.

- **Aprendizaje No Supervisado**: En el aprendizaje no supervisado, los datos no est√°n etiquetados. El modelo encuentra patrones ocultos o estructuras en los datos.

 - K-Means para segmentaci√≥n de pacientes seg√∫n caracter√≠sticas cl√≠nicas.

 - An√°lisis de Componentes Principales (PCA) para reducci√≥n de dimensionalidad en datos de laboratorio.

- **Aprendizaje Reforzado**: El aprendizaje reforzado se basa en la interacci√≥n de un agente con un entorno para maximizar una recompensa.

 - Algoritmos de IA para optimizar planes de tratamiento personalizados.

 - Control de dispositivos m√©dicos aut√≥nomos.

# <strong> üë®üèª‚Äç‚öïÔ∏èAplicaciones de la IA en Medicina </strong>

La IA ha transformado la medicina en varias √°reas clave:

- **Diagn√≥stico Asistido por Computadora:** Las redes neuronales se utilizan para analizar im√°genes m√©dicas, como radiograf√≠as y resonancias magn√©ticas, para detectar enfermedades como el c√°ncer m√°s r√°pidamente y con mayor precisi√≥n.

- **Predicci√≥n de Resultados Cl√≠nicos:** Los modelos de aprendizaje autom√°tico pueden predecir el riesgo de complicaciones o la probabilidad de recuperaci√≥n de pacientes bas√°ndose en sus historias cl√≠nicas y datos demogr√°ficos.

- **Desarrollo de Medicamentos:** La IA ayuda a identificar nuevos compuestos farmac√©uticos y a acelerar el proceso de descubrimiento de f√°rmacos mediante el an√°lisis de grandes conjuntos de datos qu√≠micos y biol√≥gicos.




# <strong> Historia y Ciclo de Vida de una Aplicaci√≥n de Aprendizaje de M√°quina </strong>

<center><img src="https://raw.githubusercontent.com/juliandariogiraldoocampo/ia_taltech/refs/heads/main/Ciclo_Vida_Aplicacion_ML.png" width=65%> </center>

El Aprendizaje de M√°quina **(ML)** ha evolucionado significativamente desde sus inicios, pasando de ser una disciplina acad√©mica a una herramienta fundamental en la industria. A lo largo de su historia, el desarrollo de aplicaciones de ML ha seguido un ciclo de vida bien definido que garantiza la eficacia y la eficiencia de los modelos. A continuaci√≥n, se detalla cada etapa de este ciclo:

<font size = 5>1. Investigaci√≥n y Planificaci√≥n </font>

Aunque la mayor√≠a de la literatura no lo considera, es √∫til incluir una fase inicial de investigaci√≥n y planificaci√≥n. Esto implica:

- Explorar el dominio,
- Identificar oportunidades potenciales.
- Definir los objetivos generales del proyecto.

<font size = 5>2. Identificaci√≥n del Problema </font>

La primera etapa implica identificar un problema o una oportunidad que pueda ser abordada mediante el aprendizaje de m√°quina. Esto requiere una comprensi√≥n profunda del negocio o del dominio en el que se est√° trabajando. Algunas preguntas clave para esta etapa incluyen:

- ¬øQu√© problema estamos tratando de resolver?
- ¬øCu√°les son los objetivos clave?
- ¬øQu√© datos est√°n disponibles para abordar este problema?

<font size = 5>3. Recolecci√≥n de Datos </font>

Una vez identificado el problema, la siguiente etapa es la recolecci√≥n de datos relevantes. Esto puede incluir datos estructurados, no estructurados o semiestructurados, dependiendo del tipo de problema. La calidad y la cantidad de los datos son cruciales para el √©xito del modelo.

- Fuentes de datos: Bases de datos, sensores, APIs, archivos CSV, etc.
- Consideraciones √©ticas: Privacidad, consentimiento, sesgo en los datos.

<font size = 5>4. Preparaci√≥n de Datos </font>

La preparaci√≥n de los datos es una de las etapas m√°s importantes y consumidoras de tiempo en el ciclo de vida de una aplicaci√≥n de ML. Incluye limpieza, transformaci√≥n y selecci√≥n de caracter√≠sticas relevantes.

- Limpieza de datos: Manejo de valores faltantes, eliminaci√≥n de datos duplicados.
- Transformaci√≥n de datos: Normalizaci√≥n, escalado, codificaci√≥n de variables categ√≥ricas.
- Selecci√≥n de caracter√≠sticas: Identificar las variables m√°s relevantes para el modelo.

<font size = 5>5. Ingenier√≠a de Modelos </font>

En esta etapa, se desarrollan y entrenan los modelos de aprendizaje de m√°quina. Esto puede incluir desde modelos simples como regresi√≥n lineal hasta modelos m√°s complejos como redes neuronales profundas.

- Selecci√≥n del algoritmo: Depende del tipo de problema (clasificaci√≥n, regresi√≥n, clustering) y de los datos disponibles.
- Entrenamiento del modelo: Uso de conjuntos de entrenamiento y validaci√≥n para ajustar los hiperpar√°metros.

<font size = 5>6. Evaluaci√≥n del Modelo </font>

La evaluaci√≥n es crucial para determinar si el modelo cumple con los objetivos establecidos. Se utilizan m√©tricas espec√≠ficas seg√∫n el tipo de problema, como precisi√≥n, recall, F1-score para clasificaci√≥n, o R-squared para regresi√≥n.

- M√©tricas de evaluaci√≥n: Selecci√≥n de las m√©tricas adecuadas para el problema.
- Validaci√≥n cruzada: T√©cnica para evaluar el rendimiento del modelo en diferentes conjuntos de datos.

<font size = 5>7. Despliegue </font>

Una vez que el modelo ha sido evaluado y ajustado, se despliega en un entorno de producci√≥n. Esto puede implicar integrarlo con aplicaciones existentes o crear una nueva interfaz para que los usuarios interact√∫en con el modelo.

- Contenerizaci√≥n: Uso de herramientas como Docker para asegurar la consistencia del entorno.
- APIs: Creaci√≥n de interfaces de programaci√≥n de aplicaciones para interactuar con el modelo.

<font size = 5>8. Mantenimiento y Actualizaci√≥n </font>

El mantenimiento es una etapa continua que asegura que el modelo siga siendo relevante y efectivo a lo largo del tiempo. Esto incluye monitorear el rendimiento del modelo, actualizarlo con nuevos datos y ajustar los par√°metros seg√∫n sea necesario.

- Monitoreo del rendimiento: Seguimiento continuo de las m√©tricas de evaluaci√≥n.
- Reentrenamiento: Actualizaci√≥n peri√≥dica del modelo con nuevos datos para mantener su precisi√≥n.

---
üîÉEste ciclo de vida es iterativo, y cada etapa puede influir en las dem√°s. Por ejemplo, problemas encontrados durante la evaluaci√≥n pueden requerir regresar a la etapa de preparaci√≥n de datos o ingenier√≠a de modelos.

## ‚úçüèª4. Frecuencia card√≠aca semanal

**Contexto:**

**Problema:**

Solicita cinco valores num√©ricos y almac√©nalos en una lista. Calcula el promedio de la frecuencia card√≠aca.

**Soluci√≥n esperada:**

Uso de listas, for, sum(), len().

- 100: Taquicardia

````python
# CAPTURA DE DATOS
lista_frecuencias = []
n = 5

for i in range(n):
    frecuencia = int(input('Ingrese el Valor de la Frecuencia: \n'))
    lista_frecuencias.append(frecuencia)
````

````python
# C√ÅLCULO DEL PROMEDIO
sumatoria = 0

for frecuencia in lista_frecuencias:
    sumatoria = sumatoria + frecuencia

cant  = len(lista_frecuencias)
promedio = sumatoria / cant

print(promedio)
````

````python
# C√ÅLCULO DEL PROMEDIO opcion 2
sumatoria = sum(lista_frecuencias)
cant = len(lista_frecuencias)
promedio = sumatoria/cant

print(promedio)
````

````python
# C√ÅLCULO DEL PROMEDIO opcion 3
import numpy as np

promedio = np.mean(lista_frecuencias)

print(promedio)
````

## ‚úçüèª5. An√°lisis de temperaturas cl√≠nicas

**Contexto:**

Un epidemi√≥logo necesita analizar la temperatura diaria de una muestra de 7 pacientes con s√≠ntomas febriles.

**Problema:**

Solicita 7 temperaturas y visualiza:

- Temperatura m√°xima
- Temperatura m√≠nima
- Promedio

**Soluci√≥n esperada:**

Uso de listas y funciones max(), min(), sum().

**Retos:**

- Imprime cu√°ntos pacientes tienen fiebre (> 37.5 ¬∞C).
- Define una temperatura de referencia e informa cu√°ntos superan este valor.

````python
# EJEMPLO DE SOLUCION - 1

lista = list(map(int, input('Introduce los n√∫meros de la lista separados por espacio: ').split()))
x = int(input('Introduce el n√∫mero de referencia (x): '))

# Filtrar y contar valores mayores que x
valores_mayores = [i for i in lista if i > x]
cantidad = len(valores_mayores)

print(f'Hay {cantidad} valores mayores que {x} en la lista.')
````

````python
# EJEMPLO DE SOLUCION - 2

import numpy as np

n = 3
lista_temp = []
cont = 0

for i in range(n):
    temp = float(input('Ingrese temperatura\n'))
    lista_temp.append(temp)


maxima = max(lista_temp)
minima = min(lista_temp)
media = np.mean(lista_temp)
cant = len(lista_temp)

for temperatura in lista_temp:
    if temperatura > 37.5:
        cont = cont + 1

print(f'Para {cant} temperaturas, el valor maximo es {maxima} y el m√≠nimo es {minima}, con una media de {media:.2f} y\nla cantidad de registros superiores a 37.5 es de {cont}')
````

## ‚úçüèª6. Diccionario de pacientes

**Contexto:**

Una sala de consulta debe almacenar y consultar informaci√≥n b√°sica de varios pacientes.

**Problema:**

Crea un diccionario por paciente con los campos: nombre, edad y diagn√≥stico. Guarda 3 pacientes en una lista. Permite buscar un paciente por nombre e imprime su informaci√≥n.

**Soluci√≥n esperada:**

Listas de diccionarios, b√∫squeda lineal.

**Retos:**

- Si el nombre no existe, imprime: ‚ÄúPaciente no encontrado‚Äù.
- Trata de resolverlo usando una lista de diccionarios o un diccionario de diccionarios

*Nota: Considera que los nombres pueden estar escritos con variaciones de may√∫sculas y min√∫sculas*

````python
# EJEMPLO DE SOLUCION - 1
# Creamos una LISTA que contiene diccionarios, uno por cada paciente

pacientes = [
    {"nombre": "Carlos", "edad": 45, "diagn√≥stico": "Hipertensi√≥n"},
    {"nombre": "Mar√≠a", "edad": 32, "diagn√≥stico": "Diabetes"},
    {"nombre": "Ana", "edad": 27, "diagn√≥stico": "Asma"}
]

# Solicitamos al usuario el nombre del paciente a buscar
nombre_buscado = input("Introduce el nombre del paciente que deseas buscar: ")

# B√∫squeda lineal del paciente por nombre
paciente_encontrado = None
for paciente in pacientes:
    if paciente["nombre"].lower() == nombre_buscado.lower():
        paciente_encontrado = paciente
        break

# Mostramos la informaci√≥n del paciente o un mensaje si no se encuentra
if paciente_encontrado:
    print(f"\nInformaci√≥n del paciente:\nNombre: {paciente_encontrado['nombre']}\nEdad: {paciente_encontrado['edad']}\nDiagn√≥stico: {paciente_encontrado['diagn√≥stico']}")
else:
    print("Paciente no encontrado.")
````

````python
# EJEMPLO DE SOLUCION - 2
# Creamos un DICCIONARIO que contiene diccionarios, uno por cada paciente

pacientes = {
    "Carlos": {"edad": 45, "diagn√≥stico": "Hipertensi√≥n"},
    "Mar√≠a": {"edad": 32, "diagn√≥stico": "Diabetes"},
    "Ana": {"edad": 27, "diagn√≥stico": "Asma"}
}

# Solicitamos al usuario el nombre del paciente a buscar
nombre_buscado = input("Introduce el nombre del paciente que deseas buscar: ")

# B√∫squeda del paciente por nombre (aprovechando la estructura de diccionario)
paciente_encontrado = pacientes.get(nombre_buscado)

# Mostramos la informaci√≥n del paciente o un mensaje si no se encuentra
if paciente_encontrado:
    print(f"\nInformaci√≥n del paciente:\nNombre: {nombre_buscado}\nEdad: {paciente_encontrado['edad']}\nDiagn√≥stico: {paciente_encontrado['diagn√≥stico']}")
else:
    print("Paciente no encontrado.")
````

## ‚úçüèªPROBLEMA EXTRA - Como importar un csv

En el siguiente ejemplo, tengo un archivo en formato <b>.csv</b> <i>(Comma Separated Values)</i> almacenado en mi computador, con la siguiente estructura pero con un tema distinto propuesto por otro compa√±ero <i>(La profe Yessica)</i>:

| Proveedor | Tipo Tela    | Precio | Descuento |
|-----------|--------------|--------|-----------|
| Corbeta   | Lemon Black  | 15900  | 0.10      |
| Coletex   | Java         | 14900  | 0.08      |
| Fabricato | Lima         | 16300  | 0.09      |

    
Que se construye asi:

    Proveedor,Tipo Tela,Precio,Descuento
    Corbeta,Lemon Black,15900,0.1
    Coletex,Java,14900,0.08
    Fabricato,Lima,16300,0.09


Para poder importar el archivo, se debe acudir al icono de la carpeta üìÅ que est√° en la barra izquierda al final. Luego en el menu que aparece seleccionar el √≠cono para subir archivo

````python
import csv

# Importar el csv
# 1Ô∏è‚É£Esto crea una LISTA DE DICCIONARIOS
with open('/content/Data_proveedores.csv',
          encoding = 'utf-8') as f:
          datos = list(csv.DictReader(f, delimiter=','))
````

````python
# 2Ô∏è‚É£Convertir la Lista de Diccionarios en un DICCIONARIO DE DICCIONARIOS

# Creo el diccionario vacio y en el ciclo lo "alimento" con cada fila o elemento de la lista
proveedores = {}
for fila in datos:
    proveedores[fila['Proveedor']] = fila
````

````python
# Imprimir cada elemento (diccionario.... del diccionario proveedores)
# üí°Recordemos que es un diccionario de diccionarios

print('                        Diccionarios en "proveedores"')
for proveedor in proveedores:
    print(proveedores[proveedor])

# Voy a imprimir esta linea para separar las dos impresiones
print('_'*100)

# Imprimir Solo los nombres de los proveedores (que en este caso son las claves)
print('                  Claves en cada diccionario de "proveedores"')
for proveedor in proveedores:
    print(proveedor)
````

````python
# Ahora podemos buscar en el diccionario usando get y la clave de cada Diccionario

busca_proveedor = input('Nombre del proveedor a buscar: ')
proveedor_encontrado = (proveedores.get(busca_proveedor))

if proveedor_encontrado:
    print(f"Proveedor: {proveedor_encontrado['Proveedor']}")
    print(f"Tipo Tela: {proveedor_encontrado['Tipo Tela']}")
    print(f"Precio: {proveedor_encontrado['Precio']}")
    print(f"Descuento: {proveedor_encontrado['Descuento']}")
    precio_final = float(proveedor_encontrado['Precio']) * (1 - float(proveedor_encontrado['Descuento']))
    print(f"Precio Final: {precio_final}")
else:
    print('No existe')
````

## ‚úçüèª9. Etiquetado de riesgo

**Problema:**

Pide al usuario edad y temperatura. Si edad > 65 y temperatura > 38.0, muestra "ALTO RIESGO", si no, muestra "RIESGO CONTROLADO".

**Soluci√≥n esperada:**

Condicional anidado con operador and.

**Reto:**

Devuelve las etiquetas como parte de un diccionario para posible reutilizaci√≥n posterior.

````python
import csv
import urllib.request
cont = 0

# URL del dataset
url = 'https://raw.githubusercontent.com/juliandariogiraldoocampo/ia_taltech/refs/heads/main/Etiquetado_Riesgo.csv'

# Descargar y leer el archivo CSV como lista de diccionarios
response = urllib.request.urlopen(url)

# Leer las lineas del archivo y decodificar los bytes a texto (UTF-8)
lines = [l.decode('utf-8-sig') for l in response.readlines()]

# Convertir el contenido de la lista (archivo csv) en una lista de diccionarios
data = list(csv.DictReader(lines))

pacientes = {}                  # Con esto creo un diccionario VAC√çO

for row in data:                # Recorro la data fila(row) por fila(row)
    pacientes[row['ID']] = row  # al diccionario en la Posici√≥n(fila) le llevo la fila entera

print(' '*30,'INFORME')
print('‚ñ™'*65)
# Calcular y mostrar la Etiqueta de RIESGO
for cada_paciente in pacientes:
    paciente = pacientes[cada_paciente]

    edad = int(paciente['Edad'])
    temperatura = float(paciente['Temperatura'])

    if edad > 65 and temperatura > 38:
        etiqueta = "ALTO RIESGO"
        cont = cont + 1
    else:
        etiqueta = "RIESGO CONTROLADO"

    paciente['Etiqueta'] = etiqueta

    print(f'Paciente {cada_paciente}, Edad {edad}, Temperatura {temperatura}¬∞C ‚Üí \t{etiqueta}')

print(f'La cantidad de Pacientes con ALTO RIESGO son {cont}')
````

<font size = 5>üí°Generar columnas relacionadas:
Si el g√©nero afecta el puntaje de memoria, puedes a√±adir l√≥gica condicional:

````python
import numpy as np
import pandas as pd

# Configurar semilla para reproductibilidad
np.random.seed(42)

# Inicializar Variables
n = 100                             # N√∫mero de Observaciones/registros
opciones = ["Masculino", "Femenino"]
prob =  [0.5, 0.5]                  # Probabilidad de cada opci√≥n. ejemplo 50%-50%

# Generar datos sint√©ticos de edad
edad = np.random.normal(70, 8, n)   # Media, Desv.Est√°ndar, N√∫mero de Observaciones

# Generar datos sint√©ticos de g√©nero
genero = np.random.choice(opciones, size=n, p=prob)

# Si el g√©nero afecta el puntaje de memoria
memoria = np.where(genero == "Masculino",      # Condici√≥n
                  np.random.normal(26, 5, n),  # Si... Hombres con media 26
                  np.random.normal(24, 5, n))  # Sino... Mujeres con media 24

# Crear el Dataframe con las listas
df_sintetico = pd.DataFrame({
    'Edad': edad,
    'Puntaje_memoria': memoria,
    'G√©nero': genero
})

df_sintetico.head()
````

<font size=4>**Uso t√≠pico:**

<font size = 5>üî∏KNNImputer - similar m√°s cercano:

¬øC√≥mo funciona el **KNNImputer**?

KNN utiliza el algoritmo de <u>**k vecinos m√°s cercanos** (k-Nearest Neighbors)</u> para imputar valores faltantes:

**Enfoque:** Rellena los valores faltantes considerando los valores de los k registros m√°s similares en el conjunto de datos.

**Funcionamiento:** Para cada observaci√≥n con un valor faltante, busca las k observaciones m√°s similares (usando las caracter√≠sticas disponibles) y usa sus valores para estimar el valor faltante.

**Ventajas:** Considera la estructura multidimensional de los datos y las relaciones entre variables.

````python
import pandas as pd
import numpy as np

# Carga de los Datos
url = 'https://github.com/juliandariogiraldoocampo/ia_taltech/raw/refs/heads/main/Dataset%201%20_%20Pima%20Indians%20diabetes%20dataset%20(PIDD)_modificado.xlsx'

# Cargar la hoja sin asumir encabezados
df_raw = pd.read_excel(url, header=None) # En caso de no querer cargar los encabezados del archivo

# Fila 5 contiene los nombres de las columnas reales (√≠ndice 5)
columnas = df_raw.iloc[5, 1:].tolist()

# Extraemos desde la fila 6 hasta la 774 (excluyendo las √∫ltimas 3 filas)
df = df_raw.iloc[6:774, 1:]  # <- NOTA: El √≠ndice 774 NO se incluye

# Asignamos nombres de columnas y reiniciamos √≠ndice
df.columns = columnas

df.reset_index(drop=True, inplace=True)

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

columnas_a_imputar = ['pregnancies',  'glucose',  'blood_pressure', 'skin_thickness', 'insulin', 'body_mass_index', 'diabetes_pedigree_function', 'age']

for col in columnas_a_imputar:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # asegurar tipo num√©rico

    df[col] = df[col].replace(0, np.nan)

df.head(10)
````

**Promedios de vecinos:** Los valores decimales largos (como 125.333333) son el resultado directo de promediar 3 valores de los vecinos m√°s cercanos.

**Similitud multidimensional:** <u>**KNNImputer no considera solo una columna a la vez**</u>, sino la similitud general basada en todas las variables disponibles. Esto significa que los "vecinos" se determinan por la cercan√≠a en un espacio multidimensional.

**Valores consistentes**: La imputaci√≥n es consistente con las relaciones entre variables. Por ejemplo, si pacientes con glucosa alta tienden a tener insulina al
ta, el algoritmo captura esta relaci√≥n.

**Preservaci√≥n de la distribuci√≥n:** Los valores imputados mantienen propiedades estad√≠sticas similares a los datos originales, sin introducir valores extremos o irrealistas.

Este enfoque es m√°s sofisticado que m√©todos de propagaci√≥n como ffill o bfill porque considera las relaciones entre m√∫ltiples variables, en lugar de  usar valores adyacentes en una secuencia.

**Contexto ideal:** Cuando es razonable asumir que un valor faltante es similar al valor que viene despu√©s.

## **<font size = 5>‚úÖ Reto 3. Normalizaci√≥n de Datos**

**üîç Introducci√≥n a la Normalizaci√≥n de Datos**

En el contexto del machine learning, la normalizaci√≥n es un paso cr√≠tico en el preprocesamiento de datos que consiste en ajustar las escalas de las variables num√©ricas para que todas compartan un rango o distribuci√≥n similar.

**¬øPor qu√© es importante?**

- **Evita sesgos en los modelos:** Algoritmos como K-Nearest Neighbors (KNN), Support Vector Machines (SVM) o Principal Component Analysis (PCA) son sensibles a las diferencias de magnitud entre variables. Por ejemplo, si una variable (como "ingresos anuales") est√° en miles y otra (como "edad") en decenas, el modelo podr√≠a sobrevalorar la primera simplemente por su escala num√©rica.

- **Mejora la convergencia:** En modelos basados en gradiente descendente (como redes neuronales), la normalizaci√≥n acelera el entrenamiento al equilibrar la velocidad de actualizaci√≥n de los pesos.

- **Facilita comparaciones**: Permite interpretar coeficientes o importancias de caracter√≠sticas de manera m√°s justa.

**Casos de uso com√∫n**

- Variables como glucosa (mg/dL), presi√≥n arterial (mmHg) e √≠ndice de masa corporal (kg/m¬≤) suelen estar en escalas muy distintas. Normalizarlas asegura que ninguna domine el modelo artificialmente.

> <font size=5>üí°Para tener en cuenta:</font><br>
> *La normalizaci√≥n es una t√©cnica fundamental para garantizar que los datos est√©n en condiciones √≥ptimas antes de entrenar modelos, especialmente cuando las variables tienen unidades o rangos dispares.*<br>
<i>**Nota:** Las t√©cnicas m√°s usadas incluyen `StandardScaler` (media=0, desviaci√≥n=1) y `MinMaxScaler` (escala [0, 1]). Elegir una u otra depende del algoritmo y la distribuci√≥n de los datos.

**üîç ¬øQu√© es la Curtosis?**

La curtosis (o kurtosis) es una medida estad√≠stica que cuantifica qu√© tan "pesadas" o ligeras son las colas de una distribuci√≥n comparada con la distribuci√≥n normal (gaussiana). Indica la presencia de valores extremos (outliers) y la concentraci√≥n de datos alrededor de la media. Se interpreta as√≠:

- **Curtosis = 0**: La distribuci√≥n tiene colas similares a la normal (mesoc√∫rtica).

- **Curtosis > 0**: Colas m√°s pesadas y pico m√°s agudo (leptoc√∫rtica ‚Üí m√°s outliers).

- **Curtosis < 0**: Colas m√°s ligeras y pico m√°s aplanado (platic√∫rtica ‚Üí menos outliers).

**ü§î ¬øPara qu√© sirve esto?**
- Identificar distribuciones no normales: Ayuda a decidir si aplicar transformaciones (ej: logar√≠tmica).

- Preparar datos para modelos: Algoritmos como SVM o KNN son sensibles a asimetr√≠a y outliers.

- Detectar anomal√≠as: Curtosis alta sugiere presencia de valores extremos.

> <font size = 6>üí°</font>*Transformamos datos no normales para que los modelos de machine learning funcionen mejor. Si los datos est√°n muy cargados a un lado o tienen valores extremos, los algoritmos pueden cometer errores o volverse lentos. <br><br>
Usamos transformaciones sencillas (como el logaritmo o la ra√≠z cuadrada) para **"suavizar" esos datos y que se parezcan m√°s a una campana de Gauss**, lo que ayuda a los modelos a aprender patrones de manera m√°s precisa.*

### <font size = 5> **Validaci√≥n de normalidad**


#### üî∏ Test de Shapiro-Wilk:

Eval√∫a si una muestra proviene de una distribuci√≥n normal.

<i><b>Uso ideal:</b> Para muestras peque√±as (t√≠picamente < 5000 datos).</i>

````python
from scipy.stats import shapiro

stat, p = shapiro(df['glucose'])
print(f'Estad√≠stico de Shapiro-Wilk: {stat:.4f}, valor p: {p:.4f}')

if p > 0.05:
    print('Distribuci√≥n normal (no se rechaza H0)')
else:
    print('Distribuci√≥n no normal (se rechaza H0)')
````

####üî∏ Test de Pearson:

Combina asimetr√≠a (skew) y curtosis (kurtosis) para evaluar normalidad.

<i><b>Uso ideal:</b> Para muestras m√°s grandes (> 20 datos, recomendable >50).</i>

````python
from scipy.stats import normaltest

stat, p = normaltest(df['glucose'])
print("p-value:", p)  # Si p < 0.05, rechaza normalidad
````

#### **Hip√≥tesis nula (H0) vs. Hip√≥tesis alternativa (H1)**

- **Hip√≥tesis nula (H0):** Es la "posici√≥n por defecto" que asumimos inicialmente.

 Ejemplo: "Los niveles de glucosa siguen una distribuci√≥n normal".

- **Hip√≥tesis alternativa (H1):** Es lo que queremos probar (lo opuesto a H0).

 Ejemplo: "Los niveles de glucosa no siguen una distribuci√≥n normal".
---
**Interpretaci√≥n clave (independiente del test):**

1. **Si la evidencia (p-value) es d√©bil (p > 0.05):**

 No podemos rechazar H0 (no hay pruebas suficientes para descartar la normalidad).

 No significa que H0 sea verdadera, solo que no hay datos contundentes en contra.

2. **Si la evidencia es fuerte (p ‚â§ 0.05):**

 Rechazamos H0 (hay suficiente apoyo para H1).

 Conclusi√≥n: Los datos se desv√≠an significativamente de la normalidad.

**Analog√≠a m√©dica:**

- H0: "El paciente est√° sano".
- H1: "El paciente tiene una condici√≥n".
- p-value: Resultado de un examen diagn√≥stico.

 - p > 0.05: "No hay hallazgos concluyentes para diagnosticar la condici√≥n".

 - p ‚â§ 0.05: "Hallazgos sugieren la presencia de la condici√≥n".

> <font size = 4>üö®**Pilas con esto:** <br>
> - <font size = 4>H0/H1 son afirmaciones sobre la poblaci√≥n, **no sobre la muestra**.
> - <font size = 4>Nunca se "prueba" H0, solo se eval√∫a si hay evidencia suficiente para rechazarla.
> - <font size = 4>¬°Esto aplica a cualquier test de hip√≥tesis, no solo a normalidad!

#### **Otro ejemplo de H0 y H1**

**Analog√≠a del Detective (Juicio Penal)**
- **H0 (Hip√≥tesis nula):** "El acusado es inocente" (como en "inocente hasta que se pruebe lo contrario").
- **H1 (Hip√≥tesis alternativa):** "El acusado es culpable".
- **p-value:** La fuerza de la evidencia (huellas, testigos, ADN).
 - Si p > 0.05: Evidencia d√©bil ‚Üí No condenas (no rechazas H0: "inocente").
 - Si p ‚â§ 0.05: Evidencia s√≥lida ‚Üí Condenas (rechazas H0: "culpable").

**<font size = 4>üö®Ojo, esto es clave:**

- Nunca pruebas la inocencia (H0), solo buscas pruebas contra ella.
- Un veredicto de "no culpable" <font size = 5>‚â†</font> "inocente", solo que no hubo pruebas suficientes.

### üìä **¬øQu√© es el p-value?**

El p-value (valor de p) es un n√∫mero entre 0 y 1 que nos dice si los datos siguen una distribuci√≥n normal (en forma de "campana de Gauss").

**Interpretaci√≥n:**

- Si p-value > 0.05: Los datos probablemente son normales ("aceptamos" la normalidad).
- Si p-value ‚â§ 0.05: Los datos no son normales (hay evidencia para rechazar la normalidad).
---
üß™ Comparaci√≥n con un Examen M√©dico
Imaginemos que estos tests son como pruebas diagn√≥sticas:

**1. Shapiro-Wilk (para muestras peque√±as, <50 datos):**
- Similar a un test r√°pido de glucosa. Es preciso con pocas muestras, pero pierde sensibilidad si hay muchos datos.
- Ejemplo: Si p = 0.03 ‚Üí "Hay evidencia de que los datos no son normales" (como un test positivo para una anomal√≠a).

**2. D'Agostino-Pearson (para muestras m√°s grandes):**
- Similar a un panel metab√≥lico completo. Eval√∫a dos aspectos:
 - Asimetr√≠a (si la curva est√° desplazada a izquierda/derecha).
 - Curtosis (si las colas son m√°s gruesas o planas que lo normal).
- Ejemplo: Si p = 0.20 ‚Üí "No hay evidencia suficiente para decir que los datos no son normales" (como un resultado negativo en un examen).

üìä **¬øPor qu√© importa en investigaci√≥n cl√≠nica?**
Muchos an√°lisis estad√≠sticos (como el t-test o ANOVA) asumen normalidad. Si el p-value ‚â§ 0.05:
- Alternativas: Usar pruebas no param√©tricas (Mann-Whitney, Kruskal-Wallis).
> Las pruebas no param√©tricas son m√©todos estad√≠sticos que no asumen una distribuci√≥n normal en los datos y se usan **cuando los datos son ordinales** o **no cumplen los requisitos de las pruebas tradicionales**.
> Ejemplos comunes:
> - Prueba de Mann-Whitney (comparar 2 grupos)
> - Prueba de Kruskal-Wallis (comparar 3+ grupos)
> - Prueba de Wilcoxon (datos pareados)
> Ventaja clave: Funcionan bien con muestras peque√±as o datos at√≠picos.
- Transformar datos: Aplicar logaritmo para normalizar.

## **<font size = 5>‚úÖ Reto 6. PCA - An√°lisis de Componentes Principales**

PCA es una t√©cnica que simplifica datos complejos al combinar variables correlacionadas en nuevos ejes ("componentes principales"), ordenados por la varianza que explican. Dico de optra forma, combina variables correlacionadas (ej: glucosa/insulina) en componentes simplificados que capturan la esencia de los datos.

**Objetivos Clave**

- **Reducir dimensiones:**

 Transformar 10+ variables en solo 2-3 componentes sin perder informaci√≥n cr√≠tica.

- **Capturar varianza:**

 El primer componente (PC1) explica la mayor variabilidad; el segundo (PC2) la siguiente, y as√≠ sucesivamente.

- **Visualizar patrones:**

 Graficar datos en 2D/3D para identificar agrupaciones naturales (ej: pacientes sanos vs. diab√©ticos).

**¬øPor qu√© usar PCA?**

- **Elimina redundancia:** Variables como glucosa e insulina (correlacionadas) se fusionan en un componente.

- **Optimiza modelos:** Menos dimensiones = menos sobreajuste + mayor velocidad.

- **Revela estructura oculta:** Agrupaciones no visibles en altas dimensiones.

 ><b><i>Ejemplo cl√≠nico:</b> <br>
 Si PC1 combina glu cosa, insulina y edad, podr√≠a interpretarse como un "factor de riesgo metab√≥lico".</i>

 > ‚ö†Ô∏è Consideraciones
 > - Escala tus datos antes de aplicar PCA (usar StandardScaler).
 > - Interpretabilidad: Los componentes son combinaciones matem√°ticas, no variables m√©dicas directas.

### üîπ Paso 1: Estandarizaci√≥n de las variables
PCA requiere que las variables est√©n en la misma escala.

````python
from sklearn.preprocessing import StandardScaler

# Selecci√≥n de variables num√©ricas
variables = ['glucose', 'blood_pressure', 'skin_thickness', 'insulin', 'body_mass_index', 'age']

# Estandarizaci√≥n
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[variables])
````

### üîπ Paso 2: Aplicar PCA

````python
from sklearn.decomposition import PCA

# Inicializar PCA para dos componentes
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Crear DataFrame con componentes
df_pca = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
````

### üîπ Paso 3: Visualizar componentes principales

````python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df_pca, x='PC1', y='PC2')
plt.title('Visualizaci√≥n PCA - 2 Componentes Principales')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} varianza)')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} varianza)')
plt.grid(True)
plt.show()
````

### üìä Paso 4: Interpretaci√≥n de varianza explicada

````python
# Porcentaje de varianza explicada por cada componente
print('Varianza explicada:', pca.explained_variance_ratio_)
````

````python
# PCA sin limitar componentes
pca_full = PCA()
pca_full.fit(X_scaled)

# Varianza explicada por cada componente
print(pca_full.explained_variance_ratio_)
````

El n√∫mero de componentes (n_components) se elige con base en cu√°nta varianza total del dataset explica cada componente principal.

````python
# PCA sin limitar componentes
pca_todos = PCA()
pca_todos.fit(X_scaled)

# Varianza explicada por cada componente
print(pca_todos.explained_variance_ratio_)
````

````python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o')
plt.xlabel('N√∫mero de Componentes')
plt.ylabel('Varianza Acumulada')
plt.title('VARIANZA EXPLICADA ACUMULADA\n')
plt.grid(True)
plt.show()
````
